# Movies csv

```python
movies= pd.read_csv("Data/movies.csv")
movies.head()
```

```python
print(movies.shape)
print(len(movies["movieId"].unique()))
## So we have movieId unique
```

```python
genre_array=[]
count=0
no_genre_movies=[]
#no_genre_list

for x,m,i in zip(movies['genres'],movies['title'],movies['movieId']):
    gen=x.split('|')
    for y in gen:
        if y=='(no genres listed)':
            no_genre_movies.append([m,i])
            count+=1
        genre_array.append(y)
# genre_array.append(gen)
# print(x)
# words = string.split(',')
# print(words)
```
