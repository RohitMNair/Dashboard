```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
movie_genres =  pd.read_csv(Path("preproc_datasets/movie_genres.csv")) 
ratings =  pd.read_csv(Path("preproc_datasets/ratings.csv")) 
```

```python
df = pd.DataFrame(columns=[i for i in range(29)]+["rating"])
df
```

```python
for j in movie_genres.genreId.values:
    print(j)
```

```python
for movie_id in movie_genres.movieId.values :
    temp_list = [0 for i in range(30)]
    avg_rating = ratings.loc[ratings.movieId == movie_id, "rating"].mean()
    temp_list[-1] = avg_rating
    for j in range(29):
        if j in movie_genres.loc[movie_genres["movieId"] == movie_id,"genreId"].values:
            temp_list[j] = 1
    temp = pd.DataFrame([temp_list],columns = [i for i in range(29)]+["rating"])
    df = pd.concat([df,temp])
```

```python
df.to_csv(Path('preproc_datasets/genres.csv'))
```

```python
df = pd.read_csv(Path("preproc_datasets/genres_influ.csv"))
print(df.shape)
df.iloc[110420:110430,:]
```

```python
df.dropna(subset=['rating'],inplace=True)
df.shape
```

```python
df.drop(['Unnamed: 0'], axis=1,inplace = True)
```

```python
df[df['rating'] == 'NaN']
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1:]
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
importance = model.feature_importances_
```

```python
y_train_pred = model.predict(X_train)
print(r2_score(y_train, y_train_pred))
print(mean_squared_error(y_train,y_train_pred))
```

```python
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
y_predict = model.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_predict))
r2_score(y_test, y_predict)
```
